```python
import tensorflow as tf
import numpy as np
import os
import cv2 as cv
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/drive') # Mount to the root directory first

import os

# Define the path to your data directory
data_dir = '/content/drive/MyDrive/facialdetection'

# Create lists to store the image paths and labels
train_images = []
train_labels = []
test_images = []
test_labels = []

# Iterate through the train and test folders
for folder in ['train', 'test']:
    folder_path = os.path.join(data_dir, folder)
    for label in os.listdir(folder_path):
        label_path = os.path.join(folder_path, label)
        for image_file in os.listdir(label_path):
            image_path = os.path.join(label_path, image_file)
            if folder == 'train':
                train_images.append(image_path)
                train_labels.append(label)
            else:
                test_images.append(image_path)
                test_labels.append(label)

# Print the number of images and labels in each set
print("Number of training images:", len(train_images))
print("Number of training labels:", len(train_labels))
print("Number of testing images:", len(test_images))
print("Number of testing labels:", len(test_labels))
```

```python
import numpy as np

# Assuming 'train_images' is a list of image file paths
# Load the images and store them in a NumPy array

# Placeholder for actual image loading logic, replace with your method
train_images_array = np.array([np.array(cv.imread(image_path)) for image_path in train_images])

# Now you can check the shape of the NumPy array
print(train_images_array.shape)
```

```python
test_images_array = np.array([np.array(cv.imread(image_path)) for image_path in test_images])
```

```python
resized_train_images_array = np.array([cv.resize(image, (224, 224)) for image in train_images_array])
print(resized_train_images_array.shape)
resized_test_images_array = np.array([cv.resize(image, (224, 224)) for image in test_images_array])
print(resized_test_images_array.shape)
```

```python
import random
random.shuffle(resized_train_images_array)
random.shuffle(resized_test_images_array)
```

```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model=tf.keras.applications.MobileNetV2()
```

```python
model.summary()
```

```python
#Tuning
base_input= model.layers[0].input
base_output=model.layers[-2].output
```

```python
base_output
```

```python
final_output=layers.Dense(128)(base_output)
final_output=layers.Activation('relu')(final_output)
final_output=layers.Dense(64)(final_output)
final_output=layers.Activation('relu')(final_output)
final_output=layers.Dense(7,activation='softmax')(final_output)
```

```python
final_output
```

```python
new_model=keras.Model(inputs=base_input,outputs=final_output)
```

```python
new_model.summary()
```

```python
new_model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
new_model.fit(resized_train_images_array, np.array(train_labels), epochs=15, batch_size=16) # Or an even smaller batch size
```
